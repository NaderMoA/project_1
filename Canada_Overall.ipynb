```python
import pandas as pd
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
```

```python
#2018
# Make a reference 
csv_path = Path("csv/2018.csv")

# Import as a DataFrame
canada2018 = pd.read_csv(csv_path, encoding = "utf-8")

# Remove unnecessary columns
canada2018_reduced = canada2018[["REF_DATE", "North American Industry Classification System (NAICS)", "VALUE"]]

# Rename colums
canada2018_renamed = canada2018_reduced.rename(columns = {
                                        "REF_DATE": "Date",
                                        "North American Industry Classification System (NAICS)": "Industry",
                                        "VALUE": "2018"
                                         })

canada2018 = canada2018_renamed.groupby('Industry')['2018'].sum().reset_index()


canada2018
```

```python
#2019
# Make a reference 
csv_path = Path("csv/2019.csv")

# Import as a DataFrame
canada2019 = pd.read_csv(csv_path, encoding="utf-8")

# Remove unnecessary columns
canada2019_reduced = canada2019[["REF_DATE", "North American Industry Classification System (NAICS)", "VALUE"]]

# Rename colums
canada2019_renamed = canada2019_reduced.rename(columns={
                                        "REF_DATE": "Date",
                                        "North American Industry Classification System (NAICS)": "Industry",
                                        "VALUE": "2019"
                                         })

canada2019 = canada2019_renamed.groupby('Industry')['2019'].sum().reset_index()

canada2019
```

```python
#2020
# Make a reference 
csv_path = Path("csv/2020.csv")

# Import as a DataFrame
canada2020 = pd.read_csv(csv_path, encoding="utf-8")

# Remove unnecessary columns
canada2020_reduced = canada2020[["REF_DATE", "North American Industry Classification System (NAICS)", "VALUE"]]

# Rename colums
canada2020_renamed = canada2020_reduced.rename(columns={
                                        "REF_DATE": "Date",
                                        "North American Industry Classification System (NAICS)": "Industry",
                                        "VALUE": "2020"
                                         })

canada2020 = canada2020_renamed.groupby('Industry')['2020'].sum().reset_index()

canada2020
```

```python
#2021
# Make a reference 
csv_path = Path("csv/2021.csv")

# Import as a DataFrame
canada2021 = pd.read_csv(csv_path, encoding="utf-8")

# Remove unnecessary columns
canada2021_reduced = canada2021[["REF_DATE", "North American Industry Classification System (NAICS)", "VALUE"]]

# Rename colums
canada2021_renamed = canada2021_reduced.rename(columns={
                                        "REF_DATE": "Date",
                                        "North American Industry Classification System (NAICS)": "Industry",
                                        "VALUE": "2021"
                                         })

canada2021 = canada2021_renamed.groupby('Industry')['2021'].sum().reset_index()

canada2021
```

```python
#2022
# Make a reference 
csv_path = Path("csv/2022.csv")

# Import as a DataFrame
canada2022 = pd.read_csv(csv_path, encoding="utf-8")

# Remove unnecessary columns
canada2022_reduced = canada2022[["REF_DATE", "North American Industry Classification System (NAICS)", "VALUE"]]

# Rename colums
canada2022_renamed = canada2022_reduced.rename(columns={
                                        "REF_DATE": "Date",
                                        "North American Industry Classification System (NAICS)": "Industry",
                                        "VALUE": "2022"
                                         })

canada2022 = canada2022_renamed.groupby('Industry')['2022'].sum().reset_index()

canada2022
```

```python
#Merge data & simplify

canada1819 = pd.merge(canada2018, canada2019, on='Industry')

canada1820 = pd.merge(canada1819, canada2020, on='Industry')

canada2122 = pd.merge(canada2021, canada2022, on='Industry')

canada_overall = pd.merge(canada1820, canada2122, on='Industry')

canada_overall = pd.DataFrame(canada_overall)

canada_overall
```

```python
# Rename rows using the mapping

row_mapping = {
    'Accommodation and food services [72]': 'Accommodation and food services',
    'Agriculture [111-112, 1100, 1151-1152]': 'Agriculture',
    'Business, building and other support services [55-56]': 'Business, building and other support services',
    'Construction [23]': 'Construction', 
    'Educational services [61]': 'Educational services',
    'Finance, insurance, real estate, rental and leasing [52-53]': 'Finance, insurance, real estate, rental and leasing',
    'Forestry, fishing, mining, quarrying, oil and gas [21, 113-114, 1153, 2100]': 'Forestry, fishing, mining, quarrying, oil and gas',
    'Health care and social assistance [62]': 'Health care and social assistance',
    'Information, culture and recreation [51, 71]': 'Information, culture and recreation',
    'Manufacturing [31-33]': 'Manufacturing',
    'Other services (except public administration) [81]': 'Other services (except public administration)',
    'Professional, scientific and technical services [54]': 'Professional, scientific and technical services',
    'Public administration [91]': 'Public administration',
    'Transportation and warehousing [48-49]': 'Transportation and warehousing', 
    'Utilities [22]': 'Utilities',
    'Wholesale and retail trade [41, 44-45]': 'Wholesale and retail trade'
    }

canada_overall['Industry'] = canada_overall['Industry'].replace(row_mapping)


canada_overall
```

```python
# Rename row & final adjustment

canada_overall = canada_overall.rename(index={
                                        0: 'Accommodation and food services',
                                        1: 'Agriculture',
                                        2: 'Business, building and other support services',
                                        3: 'Construction',
                                        4: 'Educational services',
                                        5: 'Finance, insurance, real estate, rental and leasing',
                                        6: 'Forestry, fishing, mining, quarrying, oil and gas',
                                        7: 'Health care and social assistance',
                                        8: 'Information, culture and recreation',
                                        9: 'Manufacturing',
                                        10: 'Other services (except public administration)',
                                        11: 'Professional, scientific and technical services',
                                        12: 'Public administration',
                                        13: 'Transportation and warehousing',
                                        14: 'Utilities',
                                        15: 'Wholesale and retail trade'
                                         })


canada_overall                                        
```

```python
# Devide by 12 months

canada_overall[['2018', '2019', '2020', '2021', '2022']] /= 12

canada_overall
```

```python
#Line chart

plt.figure(figsize=(12, 8))

for sector in canada_overall['Industry']:
    plt.plot(canada_overall.columns[1:], canada_overall.loc[canada_overall['Industry'] == sector, canada_overall.columns[1:]].values.flatten(), label=sector, marker='o')

plt.title('Employment by Industry')
plt.ylabel('(Thousands)')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()
```

```python
# Increaing rate

canada_overall['Increasing Rate'] = (canada_overall['2022'] - canada_overall['2018']) / canada_overall['2018'] * 100

canada_overall_ir = pd.DataFrame(canada_overall)

canada_overall_ir = canada_overall_ir[['Industry', 'Increasing Rate']]

canada_overall_ir
```

```python
# Chart

plt.figure(figsize=(10, 6))
canada_overall_ir['Increasing Rate'].sort_values().plot(kind='barh', color='orange')
plt.xlabel('Increasing Rate (%)')
plt.title('Canada Increasing Rate from 2018 to 2022')
plt.grid(axis='x')

plt.show()
```
