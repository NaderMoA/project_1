```python
import pandas as pd
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create csv path & dataframe

canada2018p = pd.DataFrame(pd.read_csv("csv/2018p.csv"))
canada2019p = pd.DataFrame(pd.read_csv("csv/2019p.csv"))
canada2020p = pd.DataFrame(pd.read_csv("csv/2020p.csv"))
canada2021p = pd.DataFrame(pd.read_csv("csv/2021p.csv"))
canada2022p = pd.DataFrame(pd.read_csv("csv/2022p.csv"))

# Reduce unnecessary columns
canada2018p_reduced = canada2018p[["GEO", "North American Industry Classification System (NAICS)", "VALUE"]]
canada2019p_reduced = canada2019p[["GEO", "North American Industry Classification System (NAICS)", "VALUE"]]
canada2020p_reduced = canada2020p[["GEO", "North American Industry Classification System (NAICS)", "VALUE"]]
canada2021p_reduced = canada2021p[["GEO", "North American Industry Classification System (NAICS)", "VALUE"]]
canada2022p_reduced = canada2022p[["GEO", "North American Industry Classification System (NAICS)", "VALUE"]]

canada2018p_reduced
```

```python
# Rename

canada2018p_renamed = canada2018p_reduced.rename(columns = {
                                        "GEO": "Province",
                                        "North American Industry Classification System (NAICS)": "Industry",
                                        "VALUE": "2018"
                                         })

canada2019p_renamed = canada2019p_reduced.rename(columns = {
                                        "GEO": "Province",
                                        "North American Industry Classification System (NAICS)": "Industry",
                                        "VALUE": "2019"
                                         })

canada2020p_renamed = canada2020p_reduced.rename(columns = {
                                        "GEO": "Province",
                                        "North American Industry Classification System (NAICS)": "Industry",
                                        "VALUE": "2020"
                                         })

canada2021p_renamed = canada2021p_reduced.rename(columns = {
                                        "GEO": "Province",
                                        "North American Industry Classification System (NAICS)": "Industry",
                                        "VALUE": "2021"
                                         })

canada2022p_renamed = canada2022p_reduced.rename(columns = {
                                        "GEO": "Province",
                                        "North American Industry Classification System (NAICS)": "Industry",
                                        "VALUE": "2022"
                                         })

canada2018p_renamed
```

```python
# Rename rows using the mapping

province_row_mapping = {
    'Accommodation and food services [72]': 'Accommodation and food services',
    'Agriculture [111-112, 1100, 1151-1152]': 'Agriculture',
    'Business, building and other support services [55-56]': 'Business, building and other support services',
    'Construction [23]': 'Construction', 
    'Educational services [61]': 'Educational services',
    'Finance, insurance, real estate, rental and leasing [52-53]': 'Finance, insurance, real estate, rental and leasing',
    'Forestry, fishing, mining, quarrying, oil and gas [21, 113-114, 1153, 2100]': 'Forestry, fishing, mining, quarrying, oil and gas',
    'Health care and social assistance [62]': 'Health care and social assistance',
    'Information, culture and recreation [51, 71]': 'Information, culture and recreation',
    'Manufacturing [31-33]': 'Manufacturing',
    'Other services (except public administration) [81]': 'Other services (except public administration)',
    'Professional, scientific and technical services [54]': 'Professional, scientific and technical services',
    'Public administration [91]': 'Public administration',
    'Transportation and warehousing [48-49]': 'Transportation and warehousing', 
    'Utilities [22]': 'Utilities',
    'Wholesale and retail trade [41, 44-45]': 'Wholesale and retail trade'
    }

canada2018p_renamed['Industry'] = canada2018p_renamed['Industry'].replace(province_row_mapping)
canada2019p_renamed['Industry'] = canada2019p_renamed['Industry'].replace(province_row_mapping)
canada2020p_renamed['Industry'] = canada2020p_renamed['Industry'].replace(province_row_mapping)
canada2021p_renamed['Industry'] = canada2021p_renamed['Industry'].replace(province_row_mapping)
canada2022p_renamed['Industry'] = canada2022p_renamed['Industry'].replace(province_row_mapping)

canada2019p_renamed.head(20)
```

```python
# Create a dictionary of DataFrames separated by province
province2018 = {province: group for province, group in canada2018p_renamed.groupby('Province', as_index=False)}
province2019 = {province: group for province, group in canada2019p_renamed.groupby('Province', as_index=False)}
province2020 = {province: group for province, group in canada2020p_renamed.groupby('Province', as_index=False)}
province2021 = {province: group for province, group in canada2021p_renamed.groupby('Province', as_index=False)}
province2022 = {province: group for province, group in canada2022p_renamed.groupby('Province', as_index=False)}

# Access individual DataFrames by province name
nl_2018 = province2018['Newfoundland and Labrador']
pe_2018 = province2018['Prince Edward Island']
ab_2018 = province2018['Alberta']
sk_2018 = province2018['Saskatchewan']
bc_2018 = province2018['British Columbia']
mb_2018 = province2018['Manitoba']
nb_2018 = province2018['New Brunswick']
ns_2018 = province2018['Nova Scotia']
qc_2018 = province2018['Quebec']
on_2018 = province2018['Ontario']

nl_2019 = province2019['Newfoundland and Labrador']
pe_2019 = province2019['Prince Edward Island']
ab_2019 = province2019['Alberta']
sk_2019 = province2019['Saskatchewan']
bc_2019 = province2019['British Columbia']
mb_2019 = province2019['Manitoba']
nb_2019 = province2019['New Brunswick']
ns_2019 = province2019['Nova Scotia']
qc_2019 = province2019['Quebec']
on_2019 = province2019['Ontario']

nl_2020 = province2020['Newfoundland and Labrador']
pe_2020 = province2020['Prince Edward Island']
ab_2020 = province2020['Alberta']
sk_2020 = province2020['Saskatchewan']
bc_2020 = province2020['British Columbia']
mb_2020 = province2020['Manitoba']
nb_2020 = province2020['New Brunswick']
ns_2020 = province2020['Nova Scotia']
qc_2020 = province2020['Quebec']
on_2020 = province2020['Ontario']

nl_2021 = province2021['Newfoundland and Labrador']
pe_2021 = province2021['Prince Edward Island']
ab_2021 = province2021['Alberta']
sk_2021 = province2021['Saskatchewan']
bc_2021 = province2021['British Columbia']
mb_2021 = province2021['Manitoba']
nb_2021 = province2021['New Brunswick']
ns_2021 = province2021['Nova Scotia']
qc_2021 = province2021['Quebec']
on_2021 = province2021['Ontario']

nl_2022 = province2022['Newfoundland and Labrador']
pe_2022 = province2022['Prince Edward Island']
ab_2022 = province2022['Alberta']
sk_2022 = province2022['Saskatchewan']
bc_2022 = province2022['British Columbia']
mb_2022 = province2022['Manitoba']
nb_2022 = province2022['New Brunswick']
ns_2022 = province2022['Nova Scotia']
qc_2022 = province2022['Quebec']
on_2022 = province2022['Ontario']
```

```python
on_1819 = pd.merge(on_2018, on_2019, on=('Province', 'Industry'))

on_1820 = pd.merge(on_1819, on_2020, on=('Province', 'Industry'))

on_2122 = pd.merge(on_2021, on_2022, on=('Province', 'Industry'))

on_1822 = pd.merge(on_1820, on_2122, on=('Province', 'Industry'))

on_overall = pd.DataFrame(on_1822)

on_overall = on_overall[['Industry', '2018', '2020', '2022']]

on_overall
```

```python
# Increaing rate

on_overall['Increasing Rate'] = (on_overall['2022'] - on_overall['2018']) / on_overall['2018'] * 100

on_overall_ir = pd.DataFrame(on_overall)

on_overall_ir = on_overall_ir[['Industry', 'Increasing Rate']]

on_overall_ir
```

```python
# Chart

on_overall_ir = on_overall_ir.sort_values(by='Increasing Rate', ascending=True)


plt.figure(figsize=(10, 6))
plt.barh(on_overall_ir['Industry'], on_overall_ir['Increasing Rate'], color='skyblue')
plt.xlabel('Increasing Rate (%)')
plt.title('Ontario')
plt.grid(axis='x')


plt.show()
```
