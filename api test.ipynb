```python
import requests
import json

url = 'https://power.larc.nasa.gov/api/application/windrose/point?start=20230101&end=20231201&latitude=43.651890&longitude=-79.381706&format=JSON'

response = requests.get(url)
if response.status_code == 200:
    data = response.json()

    # Print the JSON structure
    print(json.dumps(data, indent=4))

    # Alternatively, you can list the top-level keys
    print("Top-level keys:", data.keys())

    # If you want to go deeper, you can check for nested structures within each key
    for key in data.keys():
        if isinstance(data[key], dict):
            print(f"Nested keys in '{key}':", data[key].keys())
        elif isinstance(data[key], list):
            print(f"'{key}' contains a list with {len(data[key])} items.")
else:
    print(f"Failed to fetch data: Status code {response.status_code}")
```
